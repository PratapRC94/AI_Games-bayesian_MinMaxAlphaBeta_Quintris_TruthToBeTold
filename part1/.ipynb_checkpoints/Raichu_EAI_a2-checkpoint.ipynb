```python
def board_to_string(board, N):
    return "\n".join(board[i:i+N] for i in range(0, len(board), N))
```

```python
def find_best_move(board, N, player, timelimit):
    # This sample code just returns the same board over and over again (which
    # isn't a valid move anyway.) Replace this with your code!
    #
    while True:
        time.sleep(1)
        yield board
```

```python
def white_pichu_move(board,N):

    fringe = []
    for i in range(2*N,N*(N-1)):
        if board[i] == 'w' and board[i+N-1] == '.' and i%N!=0: # not in range(2*N,N*N,N) :
            boardc = list(copy.deepcopy(board))
            if i+N-1 in range(N*(N-1),N*N):
                boardc[i+N-1] = '@'
            else:
                boardc[i+N-1] = 'w'
            boardc[i] ='.'
            boardc = "".join(boardc)
            fringe.append(boardc)

        if (i+1)%N!=0 and board[i] == 'w' and board[i+N+1] == '.': #i not in range(3*N-1,N*N,N) :
            boardc = list(copy.deepcopy(board))
            if i+N+1 in range(N*(N-1),N*N):
                boardc[i+N+1] = '@'
            else:
                boardc[i+N+1] = 'w'
            boardc[i] ='.'
            boardc = "".join(boardc)
            fringe.append(boardc)


        if i not in range(N*(N-2),N*N):
            if board[i] == 'w' and board[i+N-1] == 'b' and board[i + 2*N - 2] == '.' and i%N not in [0,1]:
                boardc = list(copy.deepcopy(board))
                # white pichu becomes pikachu if it jumps to last row
                if i+2*N-2 in [j for j in range(N*(N-1), N*N)]:
                    boardc[i+2*N-2] = '@'
                else:
                    boardc[i+2*N-2] = 'w'
                boardc[i] = '.'
                boardc[i+N-1] = '.'
                boardc = "".join(boardc)
                fringe.append(boardc)

        if i not in range(N*(N-2),N*N):
            if board[i] == 'w' and board[i+N+1] == 'b' and board[i + 2*N + 2] == '.' and i%N not in [N-1,N-2]:
                boardc = list(copy.deepcopy(board))
                # white pichu becomes pikachu if it jumps to last row
                if i+2*N+2 in [j for j in range(N*(N-1), N*N)]:
                    boardc[i+2*N+2] = '@'
                else:
                    boardc[i+2*N+2] = 'w'
                boardc[i] = '.'
                boardc[i+N+1] = '.'
                boardc = "".join(boardc)
                fringe.append(boardc)
    return fringe
```

```python
def white_pikachu_move(board,N):
    
    fringe = []
    for i in range(N,N*(N-1)):
        # forward
        if board[i]=='W':
            for j in [i+N,i+2*N]:
                if j<N*N:
                    if board[j]!='.':
                        break   
                    if board[j]=='.':
                        boardc = list(copy.deepcopy(board))
                        if j in range(N*(N-1),N*N):
                            boardc[j]=='@'
                        else:
                            boardc[j] = 'W'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                    
        # left
        if board[i]=='W' and i%N!=0:
            for j in [i-1,i-2]:
                if (j+1)%N!=0: #and (j+1)%N!=0:
                    if board[j]!='.':
                        break   
                    if board[j]=='.':
                        boardc = list(copy.deepcopy(board))
                        boardc[j] = 'W'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                
        # right
        if board[i]=='W' and (i+1)%N!=0:
            for j in [i+1,i+2]:
                if j%N!=0:#and j%N!=0:
                    if board[j]!='.':
                        break   
                    if board[j]=='.':
                        boardc = list(copy.deepcopy(board))
                        boardc[j] = 'W'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                        
        
        # forward kill                
        if i not in range(N*(N-2),N*N) and board[i]=='W':
            for k in [i+N,i+2*N]:
                for j in [i+2*N,i+3*N]:
                    if j!=k and j<N*N :
                        if board[j]!='.': break
                        if board[j]=='.':
                            #for k in [i+N,i+2*N]:
                            if board[k]=='w' or board[k]=='W': break
                            elif board[k]=='.': continue
                            elif board[k]=='b' or board[k]=='B':
                                if k==i+2*N and board[i+N]!='.': break
                                boardc = list(copy.deepcopy(board))
                                if j in range(N*(N-1),N*N):
                                    boardc[j]='@'
                                else:
                                    boardc[j]='W'
                                boardc[k]='.'
                                boardc[i]='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                
        # left kill
        if i not in range(N*(N-2),N*(N-1)) and board[i]=='W':
            if i%N not in [0,1]: 
                for k in [i-1,i-2]:
                    for j in [i-2,i-3]:
                        if j!=k and (j+1)%N!=0 :
                            if board[j]!='.': break
                            if board[j]=='.':
                                if board[k]=='w' or board[k]=='W': break
                                elif board[k]=='.': continue
                                elif board[k]=='b' or board[k]=='B':
                                    
                                    if k==i-2 and board[i-1]!='.': break
                                    boardc = list(copy.deepcopy(board))
                                    boardc[j]='W'
                                    boardc[k]='.'
                                    boardc[i]='.'
                                    boardc = "".join(boardc)
                                    fringe.append(boardc)
                          
        # right kill
        if i not in range(N*(N-2),N*(N-1)) and board[i]=='W':
            if i%N not in [N-1,N-2]: 
                for k in [i+1,i+2]:
                    for j in [i+2,i+3]:
                        if j!=k and j%N!=0 :
                            if board[j]!='.': break
                            if board[j]=='.':
                                if board[k]=='w' or board[k]=='W': break
                                elif board[k]=='.': continue
                                elif board[k]=='b' or board[k]=='B':
                                    
                                    if k==i+2 and board[i+1]!='.': break
                                    boardc = list(copy.deepcopy(board))
                                    boardc[j]='W'
                                    boardc[k]='.'
                                    boardc[i]='.'
                                    boardc = "".join(boardc)
                                    fringe.append(boardc)
       
        
    
    return fringe
```

```python
def white_raichu_move(board,N):
    
    fringe = []
    for i in range(0,N*N):
        # forward
        if board[i]=='@':
            for j in range(i-N,-1,-N):
                if j>=0:
                    if board[j]!='.':
                        break   
                    if board[j]=='.':
                        boardc = list(copy.deepcopy(board))
                        boardc[j] = '@'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                    
        
        # diagonal left
        if board[i] == '@' and i%N!=0: #not in range(2*N,N*N,N) :
            m=i%N
            for j in range(1,m+1):
                if i-j*N-j>=0:
                    if board[i-j*N-j]!='.':
                        break 
                    if board[i-j*N-j]=='.':
                        boardc = list(copy.deepcopy(board))    
                        boardc[i-j*N-j] = '@'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                  
        # diagonal right
        if board[i] == '@' and (i+1)%N!=0: #not in range(2*N,N*N,N) :
            m = N-1-(i%N)
            for j in range(1,m+1):
                if i-j*N+j>=0:
                    if board[i-j*N+j]!='.':
                        break 
                    if board[i-j*N+j]=='.':
                        boardc = list(copy.deepcopy(board))    
                        boardc[i-j*N+j] = '@'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)

        
        
        # white raichu moving backwards
        if i not in range(N*(N-1),N*N) and board[i]=='@':
            for j in range(i+N,N*N,N):
                if board[j] != '.': break
                if board[j] == '.':
                    boardc = list(copy.deepcopy(board))
                    boardc[j] = '@'
                    boardc[i] ='.'
                    boardc = "".join(boardc)
                    fringe.append(boardc)
                    
               
        # left
        if board[i]=='@' and i%N!=0:
            m = i%N
            for j in range(1,m+1):
                if board[i-j]!='.':
                    break   
                if board[i-j]=='.':
                    boardc = list(copy.deepcopy(board))
                    boardc[i-j] = '@'
                    boardc[i] ='.'
                    boardc = "".join(boardc)
                    fringe.append(boardc)
                        
              
        # right
        if board[i]=='@' and (i+1)%N!=0:
            m = N-1-(i%N)
            for j in range(1,m+1):
                    if board[i+j]!='.':
                        break   
                    if board[i+j]=='.':
                        boardc = list(copy.deepcopy(board))
                        boardc[i+j] = '@'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                        
        
         # white raichu moving forward and jumping above black by killing it
        if i not in range(0,2*N) and board[i]=='@':
            for k in range(i-N,-1,-N):
                if board[k] == 'w' or board[k] == 'W' or board[k]=='@': break
                elif board[k]  =='.': continue
                elif board[k] =='b' or board[k] == 'B' or board[k]=='$':
                    for j in range(k-N,-1,-N):
                        if j!=k and j>=0:
                            if board[j] != '.': break
                            if board[j] == '.':
                                for l in range(2,N-1):
                                    #print(board[i-l*N],board[i-N]!='.')
                                    if k==i-l*N and (board[i-p*N]!='.' for p in range(1,l)): break
                                boardc = list(copy.deepcopy(board))
                                boardc[j] = '@'
                                boardc[i] ='.'
                                boardc[k] ='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break         
        
        # forward killing diagonal left
        if i not in range(0,2*N) and i%N not in [0,1] and board[i] == '@': #not in range(2*N,N*N,N) :
            m=i%N
            for k in range(1,m):
                if i-k*N-k>=0:
                    if board[i-k*N-k]=='w' or board[i-k*N-k]=='W' or board[i-k*N-k]=='@': break
                    elif board[i-k*N-k]=='.': continue
                    elif board[i-k*N-k]=='b' or board[i-k*N-k]=='B' or board[i-k*N-k]=='$':
                        for j in range(k+1,m+1):
                            #if j!=k and (j+1)%N!=0 :
                            if i-j*N-j>=0:
                                if board[i-j*N-j]!='.': break
                                if board[i-j*N-j]=='.':
                                    for l in range(2,m):
                                        if k==i-l*N-l and (board[i-p*N-p]!='.' for p in range(1,l)): break
                                    boardc = list(copy.deepcopy(board))
                                    boardc[i-j*N-j]='@'
                                    boardc[i-k*N-k]='.'
                                    boardc[i]='.'
                                    boardc = "".join(boardc)
                                    fringe.append(boardc)
                        break
        
        # forward killing diagonal right
        if i not in range(0,2*N) and i%N not in [N-1,N-2] and board[i] == '@': #not in range(2*N,N*N,N) :
            m=N-1-(i%N)
            for k in range(1,m):
                if i-k*N+k>=0:
                    if board[i-k*N+k]=='w' or board[i-k*N+k]=='W' or board[i-k*N+k]=='@': break
                    elif board[i-k*N+k]=='.': continue
                    elif board[i-k*N+k]=='b' or board[i-k*N+k]=='B' or board[i-k*N+k]=='$':
                        for j in range(k+1,m+1):
                            #if j!=k and (j+1)%N!=0 :
                            if i-j*N+j>=0:
                                if board[i-j*N+j]!='.': break
                                if board[i-j*N+j]=='.':
                                    for l in range(2,m):
                                        if k==i-l*N+l and (board[i-p*N+p]!='.' for p in range(1,l)): break
                                    boardc = list(copy.deepcopy(board))
                                    boardc[i-j*N+j]='@'
                                    boardc[i-k*N+k]='.'
                                    boardc[i]='.'
                                    boardc = "".join(boardc)
                                    fringe.append(boardc)
                        break
   
   
        # white raichu moving backward and jumping above black by killing it
        if i not in range(N*(N-2),N*N) and board[i]=='@':
            for k in range(i+N,N*N,N):
                if board[k] == 'w' or board[k] == 'W' or board[k]=='@': break
                elif board[k]  =='.': continue
                elif board[k] =='b' or board[k] == 'B' or board[k]=='$':
                    for j in range(k+N,N*N,N):
                        if j!=k and j<N*N:
                            if board[j]!='.': break
                            if board[j]=='.':
                                for l in range(2,N-1):
                                    if k==i+l*N and (board[i+p*N]!='.' for p in range(1,l)): break
                                boardc = list(copy.deepcopy(board))
                                boardc[j]='@'
                                boardc[k]='.'
                                boardc[i]='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break
            
                                                           
      
        # left kill
        #if i not in range(N*(N-2),N*(N-1)) and board[i]=='W':
        if board[i]=='@' and i%N not in [0,1]:
            m=i%N
            for k in range(i-1,i-m-1,-1):
                if board[k]=='w' or board[k]=='W' or board[k]=='@': break
                elif board[k]=='.': continue
                elif board[k]=='b' or board[k]=='B' or board[k]=='$':
                    for j in range(k-1,i-m-1,-1):
                        if j!=k and (j+1)%N!=0 :
                            if board[j]!='.': break
                            if board[j]=='.':
                                for l in range(2,m):
                                    if k==i-l and (board[i-p]!='.' for p in range(1,l)): break
                                boardc = list(copy.deepcopy(board))
                                boardc[j]='@'
                                boardc[k]='.'
                                boardc[i]='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break
        
              
                    
        # right kill
        #if i not in range(N*(N-2),N*(N-1)) and board[i]=='W':
        if board[i]=='@' and i%N not in [N-1,N-2]:
            m = N-1-(i%N)
            for k in range(i+1,i+m):
                if board[k]=='w' or board[k]=='W' or board[k]=='@': break
                elif board[k]=='.': continue
                elif board[k]=='b' or board[k]=='B' or board[k]=='$':
                    for j in range(k+1,i+m+1):
                        if j!=k and j%N!=0 :
                            if board[j]!='.': break
                            if board[j]=='.':
                                for l in range(2,m):
                                    if k==i+l and (board[i+p]!='.' for p in range(1,l)):break
                                boardc = list(copy.deepcopy(board))
                                boardc[j]='@'
                                boardc[k]='.'
                                boardc[i]='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break
   
    return fringe
```

```python
def black_raichu_move(board,N):
    
    fringe = []
    for i in range(0,N*N):
        # forward
        if board[i]=='$':
            for j in range(i+N,N*N,N):
                if j<N*N:
                    if board[j]!='.':
                        break   
                    if board[j]=='.':
                        boardc = list(copy.deepcopy(board))
                        boardc[j] = '$'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                    
    
        # diagonal left
        if board[i] == '$' and i%N!=0: #not in range(2*N,N*N,N) :
            m=i%N
            for j in range(1,m+1):
                if i+j*N-j<N*N:
                    if board[i+j*N-j]!='.':
                        break 
                    if board[i+j*N-j]=='.':
                        boardc = list(copy.deepcopy(board))    
                        boardc[i+j*N-j] = '$'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
                  
        # diagonal right
        if board[i] == '$' and (i+1)%N!=0: #not in range(2*N,N*N,N) :
            m = N-1-(i%N)
            for j in range(1,m+1):
                if i+j*N+j<N*N:
                    if board[i+j*N+j]!='.':
                        break 
                    if board[i+j*N+j]=='.':
                        boardc = list(copy.deepcopy(board))    
                        boardc[i+j*N+j] = '$'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)

        
       
        # black raichu moving backwards
        if i not in range(0,N) and board[i]=='$':
            for j in range(i-N,-1,-N):
                if board[j] != '.': break
                if board[j] == '.':
                    boardc = list(copy.deepcopy(board))
                    boardc[j] = '$'
                    boardc[i] ='.'
                    boardc = "".join(boardc)
                    fringe.append(boardc)
                    
               
        # left
        if board[i]=='$' and i%N!=0:
            m = i%N
            for j in range(1,m+1):
                if board[i-j]!='.':
                    break   
                if board[i-j]=='.':
                    boardc = list(copy.deepcopy(board))
                    boardc[i-j] = '$'
                    boardc[i] ='.'
                    boardc = "".join(boardc)
                    fringe.append(boardc)
                        
              
        # right
        if board[i]=='$' and (i+1)%N!=0:
            m = N-1-(i%N)
            for j in range(1,m+1):
                    if board[i+j]!='.':
                        break   
                    if board[i+j]=='.':
                        boardc = list(copy.deepcopy(board))
                        boardc[i+j] = '$'
                        boardc[i] ='.'
                        boardc = "".join(boardc)
                        fringe.append(boardc)
      
        # black raichu moving forward and jumping above white by killing it
        if i not in range(N*(N-2),N*N) and board[i]=='$':
            for k in range(i+N,N*N,N):
                if board[k] == 'b' or board[k] == 'B' or board[k]=='$': break
                elif board[k]  =='.': continue
                elif board[k] =='w' or board[k] == 'W' or board[k]=='@':
                    for j in range(k+N,N*N,N):
                        if j!=k and j<N*N:
                            if board[j] != '.': break
                            if board[j] == '.':
                                for l in range(2,N-1):
                                    #print(board[i-l*N],board[i-N]!='.')
                                    if k==i+l*N and (board[i+p*N]!='.' for p in range(1,l)): break
                                boardc = list(copy.deepcopy(board))
                                boardc[j] = '$'
                                boardc[i] ='.'
                                boardc[k] ='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break         
  
        # forward killing diagonal left
        if i not in range(N*(N-2),N*N) and i%N not in [0,1] and board[i] == '$': #not in range(2*N,N*N,N) :
            m=i%N
            for k in range(1,m):
                if i+k*N-k<N*N:
                    if board[i+k*N-k]=='b' or board[i+k*N-k]=='B' or board[i+k*N-k]=='$': break
                    elif board[i+k*N-k]=='.': continue
                    elif board[i+k*N-k]=='w' or board[i+k*N-k]=='W' or board[i+k*N-k]=='@':
                        for j in range(k+1,m+1):
                            #if j!=k and (j+1)%N!=0 :
                            if i+j*N-j<N*N:
                                if board[i+j*N-j]!='.': break
                                if board[i+j*N-j]=='.':
                                    for l in range(2,m):
                                        if k==i+l*N-l and (board[i+p*N-p]!='.' for p in range(1,l)): break
                                    boardc = list(copy.deepcopy(board))
                                    boardc[i+j*N-j]='$'
                                    boardc[i+k*N-k]='.'
                                    boardc[i]='.'
                                    boardc = "".join(boardc)
                                    fringe.append(boardc)
                        break
    
        # forward killing diagonal right
        if i not in range(N*(N-1),N*N) and i%N not in [N-1,N-2] and board[i] == '$': #not in range(2*N,N*N,N) :
            m=N-1-(i%N)
            for k in range(1,m):
                if i+k*N+k<N*N:
                    if board[i+k*N+k]=='b' or board[i+k*N+k]=='B' or board[i+k*N+k]=='$': break
                    elif board[i+k*N+k]=='.': continue
                    elif board[i+k*N+k]=='w' or board[i+k*N+k]=='W' or board[i+k*N+k]=='@':
                        for j in range(k+1,m+1):
                            #if j!=k and (j+1)%N!=0 :
                            if i+j*N+j<N*N:
                                if board[i+j*N+j]!='.': break
                                if board[i+j*N+j]=='.':
                                    for l in range(2,m):
                                        if k==i+l*N+l and (board[i+p*N+p]!='.' for p in range(1,l)): break
                                    boardc = list(copy.deepcopy(board))
                                    boardc[i+j*N+j]='$'
                                    boardc[i+k*N+k]='.'
                                    boardc[i]='.'
                                    boardc = "".join(boardc)
                                    fringe.append(boardc)
                        break
                    
        # black raichu moving backward and jumping above white by killing it
        if i not in range(0,2*N) and board[i]=='$':
            for k in range(i-N,-1,-N):
                if board[k] == 'b' or board[k] == 'B' or board[k]=='$': break
                elif board[k]  =='.': continue
                elif board[k] =='w' or board[k] == 'W' or board[k]=='@':
                    for j in range(k-N,-1,-N):
                        if j!=k and j>=0:
                            if board[j]!='.': break
                            if board[j]=='.':
                                for l in range(2,N-1):
                                    if k==i-l*N and (board[i-p*N]!='.' for p in range(1,l)): break
                                boardc = list(copy.deepcopy(board))
                                boardc[j]='$'
                                boardc[k]='.'
                                boardc[i]='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break
            
                                                           
       
        # left kill
        #if i not in range(N*(N-2),N*(N-1)) and board[i]=='W':
        if board[i]=='$' and i%N not in [0,1]:
            m=i%N
            for k in range(i-1,i-m-1,-1):
                if board[k]=='b' or board[k]=='B' or board[k]=='$': break
                elif board[k]=='.': continue
                elif board[k]=='w' or board[k]=='W' or board[k]=='@':
                    for j in range(k-1,i-m-1,-1):
                        if j!=k and (j+1)%N!=0 :
                            if board[j]!='.': break
                            if board[j]=='.':
                                for l in range(2,m):
                                    if k==i-l and (board[i-p]!='.' for p in range(1,l)): break
                                boardc = list(copy.deepcopy(board))
                                boardc[j]='$'
                                boardc[k]='.'
                                boardc[i]='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break
        
              
              
        # right kill
        #if i not in range(N*(N-2),N*(N-1)) and board[i]=='W':
        if board[i]=='$' and i%N not in [N-1,N-2]:
            m = N-1-(i%N)
            for k in range(i+1,i+m):
                if board[k]=='b' or board[k]=='B' or board[k]=='$': break
                elif board[k]=='.': continue
                elif board[k]=='w' or board[k]=='W' or board[k]=='@':
                    for j in range(k+1,i+m+1):
                        if j!=k and j%N!=0 :
                            if board[j]!='.': break
                            if board[j]=='.':
                                for l in range(2,m):
                                    if k==i+l and (board[i+p]!='.' for p in range(1,l)):break
                                boardc = list(copy.deepcopy(board))
                                boardc[j]='$'
                                boardc[k]='.'
                                boardc[i]='.'
                                boardc = "".join(boardc)
                                fringe.append(boardc)
                    break
       
    return fringe
```

```python
# evaluation function for calculating the utility of a particlar board configuration
def get_Utility(board,max_player,N):
    
    # for calculating the number of white and black pichus and pikachus on the board
    num_pichu_w = board.count('w')
    num_pichu_b = board.count('b')
    num_pikachu_w = board.count('W')
    num_pikachu_b = board.count('B')
    num_raichu_w = board.count('@')
    num_raichu_b = board.count('$')
    
    # for calculating the number of white and black pieces of pichu on the opponents half i.e white on N-1 and N-2 row and black on 1 and 2 row
    num_pichu_b_adv = board[N:3*N].count('b')
    num_pichu_w_adv = board[N*(N-3):N*(N-1)].count('w')
    num_pikachu_b_adv = board[N:3*N].count('B')
    num_pikachu_w_adv = board[N*(N-3):N*(N-1)].count('W')
    num_raichu_b_adv = board[0:N].count('$')
    num_raichu_w_adv = board[N*(N-1):N*N].count('@')
    
    # calculating the mobility of white and black pieces
    num_mob_w = len(white_pichu_move(board,N))+len(white_pikachu_move(board,N))+len(white_raichu_move(board,N))
    num_mob_b = len(black_pichu_move(board,N))+len(black_pikachu_move(board,N))+len(black_raichu_move(board,N))

    if max_player == 'w':
        return 5*(num_pichu_w-num_pichu_b)+20*(num_pikachu_w-num_pikachu_b)+25*(num_raichu_w-num_raichu_b)+10*(num_pichu_w_adv-num_pichu_b_adv)+10*(num_pikachu_w_adv-num_pikachu_b_adv)+0.5*(num_mob_w-num_mob_b)
    elif max_player == 'b':
        return 5*(num_pichu_b-num_pichu_w)+20*(num_pikachu_b-num_pikachu_w)+25*(num_raichu_b-num_raichu_w)+10*(num_pichu_b_adv-num_pichu_w_adv)+10*(num_pikachu_b_adv-num_pikachu_w_adv)+0.5*(num_mob_b-num_mob_w)
    
    
    
    
```

```python
def terminal_state(board,max_player):
    num_pichu_w = board.count('w')
    num_pichu_b = board.count('b')
    num_pikachu_w = board.count('W')
    num_pikachu_b = board.count('B')
    num_raichu_w = board.count('@')
    num_raichu_b = board.count('$')
    # for calculating the number of white and black pichus and pikachus on the board
    if (num_pichu_b+num_pikachu_b+num_raichu_b) == 0 and max_player == 'w':
        return True
    elif (num_pichu_w+num_pikachu_w+num_raichu_w) == 0 and max_player == 'b':
        return True
    else:
        return False
```

```python
def successors(board, player,N):
    #N=int(math.sqrt(len(board)))
    if player == 'w':
        return white_pichu_move(board,N)+white_pikachu_move(board,N)+white_raichu_move(board,N)
    elif player == 'b':
        return black_pichu_move(board,N)+black_pikachu_move(board,N)+black_raichu_move(board,N)
```

```python
def min_state(board, max_player, min_player, alpha, beta, depth,N):
    best_move = None
    min_val = float('inf')

    if depth == 0 or terminal_state(board, max_player):
        return board, get_Utility(board, max_player,N)

    else:
        for s in successors(board, min_player,N):
            #print(max_method(s, max_player, min_player, alpha, beta, depth - 1,N))
            v = max_state(s, max_player, min_player, alpha, beta, depth - 1,N)[1]

            v = min(v,min_val)

            if v <= alpha:
                return (board, v)
            beta = min(beta,v)
            updated_board =s
    return (updated_board,v)
```

```python
def max_state(board, max_player, min_player, alpha, beta, depth,N):
    best_move = None
    max_val = float('-inf')

    if depth == 0 or terminal_state(board, max_player):
        return (board, get_Utility(board, min_player,N))

    else:
        for s in successors(board, max_player,N):
            #print(min_method(s, max_player, min_player, alpha, beta, depth - 1,N))
            v = min_state(s, max_player, min_player, alpha, beta, depth - 1,N)[1]

            v = max(v,max_val)

            if v >= beta:
                return (board,v)

            alpha = max(alpha,v)
            updated_board = s
        return (updated_board,v)

    return (best_move, max_val)
```

```python
def alpha_beta_pruning(board,N, max_player, min_player, depth):
    alpha = float('-inf')
    beta = float('inf')
    max_val = float('-inf')
    best_move = None

    moves = []

    for s in successors(board, max_player,N):
        v = min_state(s, max_player, min_player, alpha, beta, depth,N)[1]
        if v > max_val:
            max_val = v
            best_move = s

    return best_move
```

```python
def find_best_move(board,N,player, timelimit,depth):
    
    if player == "b":
        max_player = "b"
        min_player = "w"
    else:
        max_player = "w"
        min_player = "b"

    best_config = alpha_beta_pruning(board,N,max_player, min_player,depth)
     
    return best_config
```
